```python
import numpy as np
import pandas as pd
```

```python
sub1 = pd.read_csv('../data/submussion_xiao_1.csv')
sub12 = pd.read_csv('../data/submussion_xiao_1_2.csv')
sub13 = pd.read_csv('../data/submussion_xiao_1_3.csv')
sub14 = pd.read_csv('../data/submussion_xiao_1_4.csv')
```

```python
sub12.rename(columns={'FORTARGET':'FORTARGET12','PROB':'PROB12'},inplace=True)
sub13.rename(columns={'FORTARGET':'FORTARGET13','PROB':'PROB13'},inplace=True)
sub14.rename(columns={'FORTARGET':'FORTARGET14','PROB':'PROB14'},inplace=True)

```

```python
sub = sub1

sub = pd.merge(sub,sub12,how='left',on=['EID'])
sub = pd.merge(sub,sub13,how='left',on=['EID'])
sub = pd.merge(sub,sub14,how='left',on=['EID'])
```

```python
sub.sort_values(['FORTARGET'],ascending=0)

sub.loc[~sub['FORTARGET12'].isnull(),['FORTARGET','PROB']] = (sub[~sub['FORTARGET12'].isnull()][['FORTARGET','PROB']]+sub[~sub['FORTARGET12'].isnull()][['FORTARGET','PROB']]) / 2
sub.loc[~sub['FORTARGET13'].isnull(),['FORTARGET','PROB']] = (sub[~sub['FORTARGET13'].isnull()][['FORTARGET','PROB']]+sub[~sub['FORTARGET13'].isnull()][['FORTARGET','PROB']]) / 2
sub.loc[~sub['FORTARGET14'].isnull(),['FORTARGET','PROB']] = (sub[~sub['FORTARGET14'].isnull()][['FORTARGET','PROB']]+sub[~sub['FORTARGET14'].isnull()][['FORTARGET','PROB']]) / 2

```

```python
sub.sort_values(['PROB'],ascending=0)
```

```python
sub.loc[~sub['FORTARGET12'].isnull(),'FORTARGET'] = (sub[~sub['FORTARGET12'].isnull()]['FORTARGET'] + sub[~sub['FORTARGET12'].isnull()]['FORTARGET12'])/2
sub.loc[~sub['FORTARGET12'].isnull(),'PROB'] = (sub[~sub['FORTARGET12'].isnull()]['PROB'] + sub[~sub['FORTARGET12'].isnull()]['PROB12'])/2

sub.loc[~sub['FORTARGET13'].isnull(),'FORTARGET'] = (sub[~sub['FORTARGET13'].isnull()]['FORTARGET'] + sub[~sub['FORTARGET13'].isnull()]['FORTARGET13'])/2
sub.loc[~sub['FORTARGET13'].isnull(),'PROB'] = (sub[~sub['FORTARGET13'].isnull()]['PROB'] + sub[~sub['FORTARGET13'].isnull()]['PROB13'])/2

sub.loc[~sub['FORTARGET14'].isnull(),'FORTARGET'] = (sub[~sub['FORTARGET14'].isnull()]['FORTARGET'] + sub[~sub['FORTARGET14'].isnull()]['FORTARGET14'])/2
sub.loc[~sub['FORTARGET14'].isnull(),'PROB'] = (sub[~sub['FORTARGET14'].isnull()]['PROB'] + sub[~sub['FORTARGET14'].isnull()]['PROB14'])/2

```

```python
sub.loc[sub['FORTARGET']>0,'FORTARGET'] = 1
```

```python
sub[['EID','FORTARGET','PROB']].to_csv('../data/submussion_xiao_s.csv',index=False,index_label=False)
```
